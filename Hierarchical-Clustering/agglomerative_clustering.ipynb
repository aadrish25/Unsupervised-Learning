```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df.head()
```

### I will try to group the same type of flowers

```python
# standardize the data
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
X_scaled=scale.fit_transform(iris_df)
X_scaled
```

```python
# now in order to visualize the data, we apply PCA 
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X_scaled_reduced=pca.fit_transform(X_scaled)
```

```python
X_scaled_reduced
```

```python
pca.explained_variance_
```

```python
# visualize the data
plt.scatter(X_scaled_reduced[:,0],X_scaled_reduced[:,1],c=iris.target)
plt.plot()
```

#### It is observable that there will be formation of 3 clusters

```python
# now lets apply agglomerative hierarchical clustering
# firstly lets construct a dendogram
import scipy.cluster.hierarchy as sc 

# construct a figure
plt.figure(figsize=(20,7))
plt.title("Dendograms")

sc.dendrogram(sc.linkage(X_scaled_reduced,method="ward"))
plt.plot()
```

#### It is observable from this diagram that there will be formation of 2 clusters (observing the longest vertical line which has not been intersected)

```python
# now apply agglomerative clustering
from sklearn.cluster import AgglomerativeClustering

agg_cluster=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
agg_cluster.fit(X_scaled_reduced)
```

```python
agg_cluster.labels_
```

```python
# visualize based on labels
plt.scatter(X_scaled_reduced[:,0],X_scaled_reduced[:,1],c=agg_cluster.labels_)
plt.plot()
```
