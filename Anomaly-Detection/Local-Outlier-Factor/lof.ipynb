```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_circles
X,y=make_circles(n_samples=1000,noise=0.11,factor=0.5,random_state=42)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y)
plt.plot()
```

```python
from sklearn.neighbors import LocalOutlierFactor
lof=LocalOutlierFactor(n_neighbors=15,algorithm='kd_tree',contamination=0.05,n_jobs=-1)
y_pred=lof.fit_predict(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y_pred, cmap='coolwarm', s=40, edgecolors='k')
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
